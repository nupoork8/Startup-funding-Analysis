```python
#Concept 4: Apply Functions
#Apply = do something to every row/cell
```

```python
#Convert "$180B" to 180
def clean_value(text):
    #remove $ and B
    return text.replace('$', '').replace('B','')
#apply this function to EVERY cell in valuation column
df['Valuation_Clean'] = df['Valuation'].apply(clean_value)
```

```python
# GroupBy = Pivot Table in Excel

# Example: Count companies per industry
df.groupby('Industry')['Company'].count()

# In Excel this would be:
# - Create pivot table
# - Rows: Industry
# - Values: Count of Company

# In Pandas: ONE LINE!
```
